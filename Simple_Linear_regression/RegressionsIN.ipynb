```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error

```

```python
class LR:

    def __init__(self):
        self.m = None
        self.b = None

    def fit(self,X_train,y_train):


        #needed for find slope of line(m)
        v1=0
        v2=0

        for i in range(len(X_train)):

            v1 = v1 + ((X_train[i] - X_train.mean())*(y_train[i] - y_train.mean()))
            v2 = v2 + (X_train[i] - X_train.mean())**2
        self.m = v1/v2
        self.b = y_train.mean() - (self.m * X_train.mean())


        print("slope(m)"+str(self.m))
        print("intercept(b)"+str(self.b))



    def predict(self,X_test):
        return self.m * X_test + self.b
```

```python
df=pd.read_csv('/content/placement.csv')

X=df.iloc[:,0].values
Y=df.iloc[:,1].values
```

```python

#splitting data to test model
mylr=LR()
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=2)

mylr.fit(X_train,y_train)
y_pred=mylr.predict(X_test)



print("mse error:"+str(mean_squared_error(y_test,y_pred)))


plt.scatter(X_test,y_test)
plt.plot(X_test,y_pred,color='red')
plt.show()

```

 Now using **sklearn** linearmodel




```python

from sklearn.linear_model import LinearRegression


data=pd.read_csv('/content/placement.csv')

X=data.iloc[:,0].values.reshape(-1,1)
Y=data.iloc[:,1].values.reshape(-1,1)



X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
lr=LinearRegression()


#fitting models
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)

print("slope(m)"+str(lr.coef_))
print("intercept(b)"+str(lr.intercept_))
print("mse error:"+str(mean_squared_error(y_test,y_pred)))

plt.scatter(X_test,y_test)
plt.plot(X_test,y_pred,color='red')
plt.show()
```
